```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array(range(50))
y1 = (2*np.array(x) - 25) - np.random.randint(-20, 20, 50)
y2 = (2*np.array(x) - 25)
df_plot = pd.DataFrame({'X': x, 'Y1': y1, 'Y2': y2})
df_plot.plot()
plt.show()
```

```python
df_plot.loc[:, 'Xsq'] = x**2
df_plot.plot(x = 'Xsq', y = 'Y2')
plt.show()
```

```python
df_plot.loc[:,'Ydif'] = df_plot.loc[:,'Y1'] - df_plot.loc[:,'Y2']
df_plot.loc[:,'Ydif'].plot.bar()
plt.show()
```

```python
hist_values = np.array([1] * 20 + [2] * 40 + [3] * 30 + [4] * 10)
values = np.zeros(hist_values.shape[0])
values[hist_values == 1] = 20
values[hist_values == 2] = 30 
values[hist_values == 4] = 70
values = values + np.random.randint(-15, 15, values.shape[0])
df = pd.DataFrame({'col1': hist_values, 'col2': values})
```

```python
df.plot.scatter(x = 'col1', y = 'col2')
plt.show()
```
