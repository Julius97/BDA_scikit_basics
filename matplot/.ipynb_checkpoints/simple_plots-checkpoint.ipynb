```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([(1/10) * k * math.pi for k in range(0, 50)])
y = np.sin(x)*np.exp(-x**(1/2))
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'ro')
plt.show()
```
